```python
from PIL import Image
from torchvision import transforms



path = '/share/project/cwm/xiuyu.yang/work/dev6/DiffusionAsShader/data/demo/0007/rgb/00000.jpg'

image = Image.open(path)

W, H = image.size
if W != 480 or H != 320:
    print(W, H)

    if W > H:
        T = transforms.Compose(
                    [
                        transforms.Resize(320, interpolation=transforms.InterpolationMode.BILINEAR),
                        transforms.CenterCrop([320, 480]),
                    ]
                )
    else:
        T = transforms.Compose(
                    [
                        transforms.Resize(480, interpolation=transforms.InterpolationMode.BILINEAR),
                        transforms.CenterCrop([320, 480]),
                    ]
                )
    image = T(image)

save_path = '/share/project/cwm/xiuyu.yang/work/dev6/DiffusionAsShader/data/demo/0007/rgb/00000.png'
image.save(save_path)
image.show()
```
